```python
# SPDX-FileCopyrightText: Copyright (c) 2024 NVIDIA CORPORATION & AFFILIATES. All rights reserved.
# SPDX-License-Identifier: MIT
#
# Permission is hereby granted, free of charge, to any person obtaining a
# copy of this software and associated documentation files (the "Software"),
# to deal in the Software without restriction, including without limitation
# the rights to use, copy, modify, merge, publish, distribute, sublicense,
# and/or sell copies of the Software, and to permit persons to whom the
# Software is furnished to do so, subject to the following conditions:
#
# The above copyright notice and this permission notice shall be included in
# all copies or substantial portions of the Software.
#
# THE SOFTWARE IS PROVIDED "AS IS", WITHOUT WARRANTY OF ANY KIND, EXPRESS OR
# IMPLIED, INCLUDING BUT NOT LIMITED TO THE WARRANTIES OF MERCHANTABILITY,
# FITNESS FOR A PARTICULAR PURPOSE AND NONINFRINGEMENT. IN NO EVENT SHALL
# THE AUTHORS OR COPYRIGHT HOLDERS BE LIABLE FOR ANY CLAIM, DAMAGES OR OTHER
# LIABILITY, WHETHER IN AN ACTION OF CONTRACT, TORT OR OTHERWISE, ARISING
# FROM, OUT OF OR IN CONNECTION WITH THE SOFTWARE OR THE USE OR OTHER
# DEALINGS IN THE SOFTWARE.

import folium
import pandas as pd
import numpy as np
import requests
import polyline

# ---------------------------------------------------------
# Define an icon mapping for different location types.
# icon_mapping = {
#     'DEPOT': {"icon": "home", "prefix": "fa", "color": "blue"},
#     'Restaurant': {"icon": "cutlery", "prefix": "fa", "color": "green"},
#     'Shop': {"icon": "shopping-bag", "prefix": "fa", "color": "orange"},
#     'Pickup': {"icon": "arrow-circle-o-up", "prefix": "fa", "color": "purple"},
#     'Delivery': {"icon": "arrow-circle-o-down", "prefix": "fa", "color": "red"},
#     'default': {"icon": "info-sign", "prefix": "glyphicon", "color": "gray"},
# }

# # ---------------------------------------------------------
# def get_map_by_vehicle(curr_route_df):
#     """Generates a Folium map for the route given in curr_route_df."""
#     center_lat = curr_route_df["Latitude"].mean()
#     center_lon = curr_route_df["Longitude"].mean()
#     m = folium.Map(location=[center_lat, center_lon], zoom_start=13)
#     for idx, row in curr_route_df.iterrows():
#         # Get type from 'order_type' column
#         type_key = row.get("order_type", "default")
#         icon_opts = icon_mapping.get(type_key, icon_mapping["default"])
#         folium.Marker(
#             location=[row["Latitude"], row["Longitude"]],
#             popup=f"{row['name']} (Stop {row['stop_index']})",
#             tooltip=f"{row['name']}",
#             icon=folium.Icon(**icon_opts)
#         ).add_to(m)
#     # Draw route polyline
#     coords = curr_route_df[["Latitude", "Longitude"]].values.tolist()
#     folium.PolyLine(coords, color="blue", weight=3, opacity=0.7).add_to(m)
#     return m

# def get_combined_map(vehicle_data, locations_df, selected_vehicle_ids):
#     """Generate a combined Folium map with routes for all selected vehicles."""
#     base_map = folium.Map(location=[locations_df["Latitude"].mean(), locations_df["Longitude"].mean()], zoom_start=12)
#     # Use a series of colors for polyline rendering
#     colors = ['red','blue','green','purple','orange','darkred','lightred','beige','darkblue','darkgreen']
    
#     for i, vehicle_id in enumerate(selected_vehicle_ids):
#         v_info = solver_resp["vehicle_data"].get(vehicle_id)
#         if not v_info:
#             continue
#         route_indices = v_info.get("route", [])
#         if len(route_indices) < 2:
#             route_indices = route_indices * 2  # ensure at least two stops
#         route_df = pd.DataFrame(route_indices, columns=["stop_index"])
#         merged_df = pd.merge(route_df, locations_df, how="left", left_on="stop_index", right_index=True)
        
#         # Draw markers with custom icons
#         for j, row in merged_df.iterrows():
#             cat = row.get("order_type", "default")
#             icon_opts = icon_mapping.get(cat, icon_mapping["default"])
#             folium.Marker(
#                 location=[row["Latitude"], row["Longitude"]],
#                 popup=f"{row['name']} (Stop {row['stop_index']})",
#                 tooltip=f"{row['name']}",
#                 icon=folium.Icon(**icon_opts)
#             ).add_to(base_map)
        
#         # Draw a polyline with a distinct color for this vehicle's route.
#         poly_color = colors[i % len(colors)]
#         coords = merged_df[["Latitude", "Longitude"]].values.tolist()
#         folium.PolyLine(coords, color=poly_color, weight=3, opacity=0.7).add_to(base_map)
        
    

# def plot_order_locations(df, pdp):
#     # Set up the colours based on well's purpose
#     purpose_colour = {'DEPOT':'red', 'Restaurant':'green', 'Retailer':'blue', 'Business': 'orange'}
#     if pdp==1:
#         purpose_colour = {'DEPOT':'gray', 'Pickup':'green', 'Delivery':'blue'}

        
#     map = folium.Map(location=[df.Latitude.mean(), df.Longitude.mean()], 
#                      zoom_start=10, control_scale=True)

#     for i,row in df.iterrows():
        
#         try:
#             icon_color = purpose_colour[row['order_type']]
#         except:
#             #Catch nans
#             icon_color = 'gray'

#         #Add each row to the map
#         folium.Marker(location=(row['Latitude'],row['Longitude']),
#                       #popup = popup, 
#                       icon=folium.Icon(color=icon_color, icon='')).add_to(map)


#     return map

# def get_map_by_vehicle(curr_route_df):
    
    
#     curr_lat_lon_coords = curr_route_df[["Latitude","Longitude"]].values.tolist()
#     location_type = curr_route_df[['order_type']].to_dict()
#     m = get_map(curr_lat_lon_coords, location_type)
    
#     return m

# def get_map(my_lat_longs, location_type):
#     m = folium.Map(location=my_lat_longs[1],
#                    zoom_start=12)
#     count = 0 
    
#     purpose_colour = {'DEPOT':'gray', 'Pickup':'green', 'Delivery':'blue'}
    
#     for src_idx in range(len(my_lat_longs))[:-1]:
#         dst_idx = src_idx + 1
        
#         source = my_lat_longs[src_idx]
#         destination = my_lat_longs[dst_idx]
        
#         route = get_route(source[1], source[0], destination[1], destination[0])
        
#         folium.PolyLine(
#                 route['route'],
#                 weight=5,
#                 color='blue',
#                 opacity=0.6
#             ).add_to(m)
        
#         if src_idx == 0:
#             folium.Marker(
#                 location=[my_lat_longs[src_idx][0],my_lat_longs[src_idx][1]],
#                 icon=folium.Icon(color="red", prefix='fa')
#             ).add_to(m)
            
#         else:
#             """
#             folium.Marker(
#                 location=route['start_point'],
#                 icon=folium.Icon(color="blue",icon="fa-map-pin", prefix='fa')
#             ).add_to(m)
#             """
#             folium.Marker(
#                 location=[my_lat_longs[src_idx][0],my_lat_longs[src_idx][1]],
#                 icon=folium.Icon(color=purpose_colour[location_type['order_type'][src_idx]], prefix='fa')
#                 # icon=plugins.BeautifyIcon(
#                 #          #icon="arrow-down", 
#                 #          icon_shape="marker",
#                 #          #number=src_idx,
#                 #          color= purpose_colour[location_type['order_type'][src_id]],
#                 #      )
#             ).add_to(m)

#     return m

# def get_combined_map(vehicle_data, locations_df):
#     """Generate a combined Folium map with routes for all vehicles."""
#     # Use a standard list of distinct colors for different vehicles
#     colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 
#               'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'pink',
#               'lightblue', 'lightgreen']
    
#     # Create a base map centered at the average location of all stops
#     m = folium.Map(location=[locations_df["Latitude"].mean(), locations_df["Longitude"].mean()], zoom_start=12)
    
#     for i, vehicle_id in enumerate(vehicle_data.keys()):
#         v_info = vehicle_data[vehicle_id]
#         route_indices = v_info.get("route", [])
#         # If only one stop is present, duplicate it so we can form a route.
#         if len(route_indices) < 2:
#             route_indices = route_indices * 2
        
#         route_df = pd.DataFrame(route_indices, columns=["stop_index"])
#         merged_df = pd.merge(route_df, locations_df, how="left", left_on="stop_index", right_index=True)
        
#         # Add markers for each stop with a vehicle-specific color
#         for j, row in merged_df.iterrows():
#             folium.Marker(
#                 location=[row["Latitude"], row["Longitude"]],
#                 popup=row["name"],
#                 icon=folium.Icon(color=colors[i % len(colors)])
#             ).add_to(m)
        
#         # Draw a polyline for this vehicle's route
#         coords = merged_df[["Latitude", "Longitude"]].values.tolist()
#         folium.PolyLine(coords, color=colors[i % len(colors)], weight=3, opacity=0.7).add_to(m)
    
#     return m

# def get_route(source_long, source_lat, dest_long, dest_lat):
#     loc = "{},{};{},{}".format(source_long, source_lat, dest_long, dest_lat)
#     url = "http://router.project-osrm.org/route/v1/driving/"
#     r = requests.get(url + loc) 

#     res = r.json()   
#     routes = polyline.decode(res['routes'][0]['geometry'])
#     start_point = [res['waypoints'][0]['location'][1], res['waypoints'][0]['location'][0]]
#     end_point = [res['waypoints'][1]['location'][1], res['waypoints'][1]['location'][0]]
#     distance = res['routes'][0]['distance']
    
#     out = {'route':routes,
#         'start_point':start_point,
#         'end_point':end_point,
#         'distance':distance
#     }
    
#     return out

### updated code with icons

import folium
import pandas as pd
import requests
import polyline

# Define an icon mapping dictionary for different order types
icon_mapping = {
    'DEPOT': {"color": "blue", "icon": "home", "prefix": "fa"},
    'Pickup': {"color": "green", "icon": "shopping-bag", "prefix": "fa"},
    'Delivery': {"color": "red", "icon": "arrow-circle-down", "prefix": "fa"},
    'Restaurant': {"color": "orange", "icon": "cutlery", "prefix": "fa"},
    'Shop': {"color": "purple", "icon": "shopping-bag", "prefix": "fa"},
    'default': {"color": "gray", "icon": "info-sign", "prefix": "glyphicon"}
}

def get_map_by_vehicle(curr_route_df):
    """
    Given a DataFrame (curr_route_df) that includes columns:
     - 'Latitude', 'Longitude', 'order_type', and a 'stop_index',
    this function creates and returns a Folium map using get_map().
    """
    # Extract a list of order types from the DataFrame.
    location_types = curr_route_df['order_type'].tolist()
    my_lat_longs = curr_route_df[["Latitude", "Longitude"]].values.tolist()
    
    m = get_map(my_lat_longs, location_types)
    return m

def get_map(my_lat_longs, location_types):
    """
    Given a list of [Latitude, Longitude] pairs and a list of location types (as strings),
    this function creates a Folium map and plots the route.
    """
    # Use the second coordinate as the map center (adjust if needed)
    m = folium.Map(location=my_lat_longs[1], zoom_start=12)
    
    # Loop through each pair of consecutive stops
    for src_idx in range(len(my_lat_longs) - 1):
        dst_idx = src_idx + 1
        
        source = my_lat_longs[src_idx]
        destination = my_lat_longs[dst_idx]
        
        # Get routing between this pair from OSRM (or your chosen routing service)
        route = get_route(source[1], source[0], destination[1], destination[0])
        
        # Draw the route as a polyline
        folium.PolyLine(
            route['route'],
            weight=5,
            color='blue',
            opacity=0.6
        ).add_to(m)
        
        # Select marker icon based on the location type of the current stop.
        # For the first stop (usually the depot), we can force a red marker.
        if src_idx == 0:
            marker_opts = {"color": "red", "icon": "home", "prefix": "fa"}
        else:
            # Get the order_type for this stop from the provided list.
            # If no type is found, use the default.
            order_type = location_types[src_idx] if src_idx < len(location_types) else "default"
            marker_opts = icon_mapping.get(order_type, icon_mapping["default"])
        
        folium.Marker(
            location=[source[0], source[1]],
            icon=folium.Icon(**marker_opts)
        ).add_to(m)
        
    # Add a marker for the last stop
    last = my_lat_longs[-1]
    # For last stop, use default marker (or adjust as needed)
    folium.Marker(
        location=[last[0], last[1]],
        icon=folium.Icon(**icon_mapping.get("default"))
    ).add_to(m)
    
    return m

def get_combined_map(vehicle_data, locations_df):
    """
    Generate a combined Folium map plotting the routes for all vehicles in vehicle_data.
    `vehicle_data` is a dict with keys as vehicle IDs and values containing a 'route'
    (a list of stop indices). `locations_df` is a DataFrame with the full location details.
    """
    # Use a standard list of colors for polyline drawing: this is for differentiating vehicle routes.
    colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 
              'beige', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'pink',
              'lightblue', 'lightgreen']
    
    # Base map centered at average location of all stops in locations_df
    m = folium.Map(location=[locations_df["Latitude"].mean(), locations_df["Longitude"].mean()], zoom_start=12)
    
    # Loop over each vehicle in the vehicle_data dictionary
    for i, vehicle_id in enumerate(vehicle_data.keys()):
        v_info = vehicle_data[vehicle_id]
        route_indices = v_info.get("route", [])
        # If only one stop is present, duplicate it so that a route can be drawn.
        if len(route_indices) < 2:
            route_indices = route_indices * 2
            
        # Create a DataFrame for this vehicle's route stops.
        route_df = pd.DataFrame(route_indices, columns=["stop_index"])
        # Merge the route with locations_df (using stop_index as the key; ensure locations_df index aligns)
        merged_df = pd.merge(route_df, locations_df, how="left", left_on="stop_index", right_index=True)
        
        # Add markers for each stop along the vehicle route.
        for j, row in merged_df.iterrows():
            # Get location type for customizing icon:
            order_type = row.get("order_type", "default")
            marker_opts = icon_mapping.get(order_type, icon_mapping["default"])
            folium.Marker(
                location=[row["Latitude"], row["Longitude"]],
                popup=row["name"],
                icon=folium.Icon(**marker_opts)
            ).add_to(m)
        
        # Draw a polyline for this vehicle's route using a distinct color.
        poly_color = colors[i % len(colors)]
        coords = merged_df[["Latitude", "Longitude"]].values.tolist()
        folium.PolyLine(coords, color=poly_color, weight=3, opacity=0.7).add_to(m)
    
    return m

def get_route(source_long, source_lat, dest_long, dest_lat):
    """
    Queries the OSRM API for a route between two geographic points.
    Returns a dictionary with the route polyline, start and end points, and distance.
    """
    # Build the location string in the format: "source_long,source_lat;dest_long,dest_lat"
    loc = "{},{};{},{}".format(source_long, source_lat, dest_long, dest_lat)
    url = "http://router.project-osrm.org/route/v1/driving/"
    r = requests.get(url + loc)
    res = r.json()
    
    # Decode the polyline geometry from OSRM's response
    routes = polyline.decode(res['routes'][0]['geometry'])
    start_point = [res['waypoints'][0]['location'][1], res['waypoints'][0]['location'][0]]
    end_point = [res['waypoints'][1]['location'][1], res['waypoints'][1]['location'][0]]
    distance = res['routes'][0]['distance']
    
    out = {
      'route': routes,
      'start_point': start_point,
      'end_point': end_point,
      'distance': distance
    }
    return out
```

# Create Cost Matrices

The <code style="background:lightgreen;color:black">cost_matrix</code> models the cost between each pair of locations.  It is used by cuOpt to compute the cost of traveling from any location to any other. The cost matrix needs to be a square matrix of dimension equal to the total number of locations which inlcludes both depots and orders. In this Vehicle Routing Problem, our cost metric is travel time. This is cost we want to minimize. 

To build a a cost matrix of live traffic data, we need to use a third party map data provider. In this workflow, the cost matrix will calculate the travel time in minutes between each two pairs of locations which we build using OSRM. 

In this PDP use case we have a <code style="background:lightgreen;color:black">mixed_fleet</code> consisting of two different types of vehicles: cars and bikes. We can imagine that these two different types of vehicles have different travel time in the crowded streets of NYC. A car is more likely to be stuck in traffic whereas a bike can bypass it. We create a different cost matrix for each vehicle type

```python
def build_time_cost_matrix(df, vehicle_type):
    latitude = df["Latitude"].to_numpy()
    longitude = df["Longitude"].to_numpy()
    
    locations=""
    n_orders = len(df)
    for i in range(n_orders):
        locations = locations + "{},{};".format(longitude[i], latitude[i])
    r = requests.get("http://router.project-osrm.org/table/v1/" + vehicle_type + "/" + locations[:-1])
    routes = json.loads(r.content)
    
    # OSRM returns duration in seconds. Here we are converting to minutes
    for i in routes['durations']:
        i[:] = [x / 60 for x in i]
    
    coords_index = { i: (latitude[i], longitude[i]) for i in range(df.shape[0])}
    time_matrix = pd.DataFrame(routes['durations'])
    
    return time_matrix
```

```python
vehicles_df['vehicle_start_in_minutes'] = vehicles_df['vehicle_start'].apply(get_minutes_from_datetime)
vehicles_df['vehicle_end_in_minutes'] = vehicles_df['vehicle_end'].apply(get_minutes_from_datetime)

# Earliest a vehicle can start 
v_tw_earliest = vehicles_df['vehicle_start_in_minutes']

# Latest a vehicle will be working
v_tw_latest = vehicles_df['vehicle_end_in_minutes']

vehicle_time_windows = list(zip(v_tw_earliest, v_tw_latest)) 
```

<code style="background:lightgreen;color:black">capacities</code> is a list of how many orders each vehicle can carry in weight. In this example, a car can carry 800 lb and a bike can carry 50 lb.

```python
vehicle_capacity = [vehicles_df['vehicle_capacity'].values.tolist()]
```

<code style="background:lightgreen;color:black">vehicle_types</code> is a list that indicates whether each vehicle is a car or a bike. 

```python
vehicle_type_map = {"car":0, "bike":1}
vehicle_types = vehicles_df["vehicle_type"].replace(vehicle_type_map).tolist()
```
